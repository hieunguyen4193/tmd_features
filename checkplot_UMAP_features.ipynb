```python
import pandas as pd 
import umap
import seaborn as sns
import matplotlib.pyplot as plt
org_df = pd.read_excel("Hieu UMAP.xlsx")
features = ['CNA', 'FLEN', 'MOTIF', 'NFT', 'RATIO', 'SCNA']
for feat in features:
    print(f"working on feature {feat}")
    df = org_df[["SampleID", "Label"] + [item for item in org_df.columns if feat in item]]
    reducer = umap.UMAP()
    embedding = reducer.fit_transform(df.drop(["SampleID", "Label"], axis = 1))
    embeddingdf = pd.DataFrame(embedding, columns = ["UMAP1", "UMAP2"])
    embeddingdf["Label"] = df["Label"].values

    plt.figure(figsize=(10, 6))
    sns.scatterplot(data = embeddingdf, x = "UMAP1", y = "UMAP2", hue = "Label")
    plt.title(f"Feature {feat}")
    plt.tight_layout()
    plt.savefig(f"feat_{feat}.svg", format='svg')
    plt.show()
```
